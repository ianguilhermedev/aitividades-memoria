## FAZENDO MEMORIA SIMPLES 16x8

```python
from math import log2
TOTAL_LINHAS = 16 
TOTAL_COLUNAS = 8 

memoria = [["0" for _ in range(TOTAL_COLUNAS)] for _ in range(TOTAL_LINHAS)] 

print( 'MEMORIA 16x8' )
keepRunning = True

while keepRunning:
    print("Digite W para escrever, R para ler, L para listar e qualquer tecla para parar")
    acao = input().upper()
    if acao=="W": 
        enderecamento = int(input(f"Digite o endereço de {int(log2(TOTAL_LINHAS))} bits: "), 2)
        dado = list(input(f"Digite o dado de {TOTAL_COLUNAS} bits: "))
        memoria[enderecamento] = dado

    elif acao=="R":
        enderecamento = int(input(f"Digite o endereço de {log2(TOTAL_LINHAS)} bits: "), 2)
        print("".join(memoria[enderecamento]))

    elif acao=="L":
        for endereco in memoria:
            print("".join(endereco))

    else:
        keepRunning = False
```
