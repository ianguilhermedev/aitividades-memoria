### Questão 14



1. *MODELANDO O CUSTO DE CADA TECNOLOGIA*: Seja N o número de bits da memória. O custo total de cada solução é dado pela soma do custo da interface com o custo por bit:
    - Custo SRAM = R$1,00 + N × R$0,00002
    - Custo DRAM = R$5,00 + N × R$0,00001

2. *ENCONTRANDO O PONTO DE EQUILÍBRIO*: Para que a DRAM seja mais barata, deve-se ter Custo_DRAM < Custo_SRAM:
    - 5,00 + N × 0,00001 < 1,00 + N × 0,00002
    - 4,00 < N × 0,00001
    - N > 4,00 / 0,00001 = **400.000 bits**

   ***Portanto, a DRAM só se torna economicamente vantajosa quando a memória possui mais de 400.000 bits (≈ 400K bits). Abaixo desse valor, a SRAM é mais barata.***

    - Tempo de acesso processador → cache (T_c): 8 ns
    - Tempo de acesso cache → memória principal (T_m): 70 ns
    - Eficiência da cache (taxa de acerto): 96% → 0,96
    - Taxa de falha: 4% → 0,04

2. *CALCULANDO COM 100 ACESSOS CONSECUTIVOS*:
    - **96 acessos com acerto** (dado encontrado na cache): 96 × 8 ns = **768 ns**
    - **4 acessos com falha** (cache + memória principal): 4 × (8 + 70) ns = 4 × 78 ns = **312 ns**
    - Tempo total = 768 + 312 = **1080 ns**

3. *TEMPO MÉDIO DE ACESSO*: T_med = 1080 / 100 = **10,8 ns**

   ***Portanto, o tempo médio de acesso do sistema é de 10,8 ns.***

### Questão 33



**A afirmação NÃO é válida.**

A capacidade de armazenar programas depende exclusivamente da **capacidade da memória** (RAM e armazenamento secundário), não do poder de processamento da UCP. O processador determina a velocidade com que as instruções são executadas, mas não a quantidade de dados ou programas que podem ser armazenados. Um computador pode ter um processador extremamente rápido com pouca memória disponível, e vice-versa — são características independentes do sistema.

### Questão 34



**Não valida técnicamente.**

Aumentar a capacidade da memória principal **não torna o acesso ao disco mais rápido** — a velocidade de uma leitura/escrita no disco (tempo de busca, latência rotacional, taxa de transferência) é determinada pelo hardware do próprio dispositivo, independente da quantidade de RAM.

O que ocorre indiretamente é que uma memória principal maior **reduz a frequência** de acessos ao disco:
- Menos faltas de página (page faults), reduzindo o uso de memória virtual em disco.
- Maior espaço para cache do sistema de arquivos, mantendo mais dados acessados frequentemente em RAM.

***Portanto, mais RAM melhora o desempenho geral do sistema ao minimizar os acessos à memória secundária — mas não altera a velocidade intrínseca do acesso ao disco quando ele de fato ocorre.***
