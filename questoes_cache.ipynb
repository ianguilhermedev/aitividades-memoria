#### SLIDE 43 - Exemplos 5.2/5.3/5.4

- Dados: Capacida da Cache = 32KB, Capacidade da MP = 16MB

    1. Total de Bits da Cache = Bits da Tag + Bits da Memória
    2. Encontrando bits da memória:
        - Como a capacidade é 32KB, logo para saber a quantidade de bits é só tranformar essa quantidade em bites.
        - 32KB x 8 = 2^8K bits, portanto são necessários 2^18 bits para representar os dados na cache.

    3. Encontrando bits da tag:
        - Como a MP tem 2^24 celulas, logo chegamos a seguinte conclusão:

            - SEPARANDO: do total de 24 bits
                - como cada linha tem 8 celulas (1 bloco) de largura, logo temos que ter 3 bits para localizar essa célula
                - cada linha ocupa 2^12  (2^15KB/2^3B), logo precisa de 12 bits para representa-lo
                - Como no total são necessários 24 bits para representar tudo, sobram 9 bits para representar a tag.

            - Calculando o total de bits: 
                - Para armazenar dados já identificamos que precisamos de 2^18 bits
                - Para armazenar as tags precisamos de 9 x 2^12, ou seja:
                    - ***O total de bits é dado por 9 x 2^12 + 2^18 = 299008 bits***

_- Nota pessoal: 16 bytes é a mesma coisa que 16 células ( preferencia pois foi ensinado dessa forma em sala )_

- Encontrando o endereço divido na forma (tag|linhas|célula):
    1. MEMORIA PRINCIPAL:
        - Se a MP tem 64MB, ela precisa de 64M endereços para representar 1 célula. Logo 64M pode ser escrito como 2^26, obtendo que precisamos de 26 bits para representar cada célula.
    2. LINHAS:
        - Como a questão diz que a cache tem 2K linhas, logo isso pode ser escrito como 2^11, o que representa 11 bits para sua identificação.
    3. CELULAS EM SI:
        - A questão diz que a largura de cada linha da cache é 16 células, logo precisa de 4 bits para ser representada. 
    4. TAG:
        - Como já vimos que o total é 26 bits, 11 sendo ocupados pelo endereçamento das linhas e 4 da tag, temos que:
            - 26 = tag + 11(linhas) + 4(celula) -> tag = 11 bites

    5. CONCLUSÃO:
        - O formato do endereço é: ***[ 11bits(tag) | 11bits(linha) | 4bits(célula) ]***

Notas: bloco -> 32 células ( cada bloco corresponde a uma linha )

- identificando formato do endereço em binário:
    1. NUM. CÉLULAS:
        - Como temos 32 células por bloco, precisamos de 5 bits para representar esse número.

    2. LINHAS:
        - Temos que 128KB é o total de bytes da cache, e que cada linha contém 32B, ou seja o numero de células é dado por:
            - 128KB / 32B -> 2^12, ou seja, são necessários 12 bits para representar a linha.

    3. TAG:
        - Como a memória hexadecimal tem 7 algarismos e cada um vale 2^4, temos um total de (2^4)^7, logo 2^28, o que precisa de 28 bits para ser representado.
            - Portanto, como 28 é o total, reduzindo 12 das linhas e 5 das células, temos que a tag ocupa 11 bits desse endereço.

    4. CONCLUSÃO: 
        - Transformando de hexadecimal para binário para poder analisar o endereço:
            - 3FC92B6 -> 0011 1111 1100 1001 0010 1011 0110

        - ***Linha é representada pelos 12 bits após a tag: [0 1001 0010 101]***

#### SLIDE 49 - Exemplos 5.5/5.6/5.7

- Dados: Capacidade da Cache = 32KB, Capacidade da MP = 16MB, mapeamento = associativo completo

    1. Total de Bits da Cache = Bits dos Dados + Bits da Tag

    2. Encontrando os bits dos dados:
        - É só converter a capacidade da cache em bits.
        - 32KB x 8 = 2^15 x 2^3 = 2^18 bits, portanto são 2^18 bits para os dados (igual ao caso direto: os dados não dependem do mapeamento).

    3. Encontrando os bits da tag:
        - Como a MP tem 16MB = 2^24 células, o endereço tem 24 bits. Portanto:

            - SEPARANDO: do total de 24 bits (no associativo o formato é só TAG | CÉLULA, sem campo de linha)
                - como cada linha tem 8 células (1 bloco) de largura, são 3 bits para localizar a célula
                - como o bloco pode ir para QUALQUER linha, não há campo de linha; a tag é o número inteiro do bloco
                - sobram, então, 24 - 3 = 21 bits para a tag

            - Calculando o total de bits:
                - Para os dados já identificamos 2^18 bits
                - Para as tags precisamos de 21 x 2^12, pois cada uma das 2^12 linhas (32KB / 8B) guarda uma tag, ou seja:
                    - ***O total de bits é dado por 21 x 2^12 + 2^18 = 86.016 + 262.144 = 348.160 bits = 340 Kbits***

_dados: uso do número de linhas da cache não foi necessário ( obviamente, pois eles nem entram no formato do endereço no mapeamento associativo )_

- Encontrando o endereço divido na forma (bloco|célula):
    1. MEMORIA PRINCIPAL:
        - Se a MP tem 64MB, ela precisa de 64M endereços para representar 1 célula. Logo 64M pode ser escrito como 2^26, obtendo que precisamos de 26 bits para representar cada célula. Nesse caso ele também tem 64MB / 16B = 2^22 blocos, logo 22 bits são para representar os blocos
    
    2. CONCLUSÃO:
        - Se temos no total 26 bits de endereço e 22 são para o bloco, logo o formato do endereço é: ***[ 22bits(bloco) | 4bits(célula) ]***

_repetindo nota do 5.4: bloco -> 32 células ( cada bloco corresponde a uma linha )_

- identificando formato do endereço em binário:
    1. NUM. CÉLULAS:
        - Como temos 32 células por bloco, precisamos de 5 bits para representar esse número.

    2. BLOCOS: 
        - Como cada o endereço foi escrito por 7 algarismos hexadecimais, logo são necessários 28 bits de endereçamento, como 5 já foram utilizados para definir qual era a célula do bloco. Então sobram 23 para determinar o bloco em si.

    4. CONCLUSÃO: 
        - Transformando de hexadecimal para binário para poder analisar o endereço:
            - 3FC92B6 -> 0011 1111 1100 1001 0010 1011 0110

        - ***o bloco é representado pelos 23 bits mais significantes: [0011 1111 1100 1001 0010 101]***

#### SLIDE 58 - Exemplos 5.8/5.9/5.10

- Dados: Capacidade da Cache = 32KB, Capacidade da MP = 16MB, conjunto de 4, linha de 8 bytes

    1. Total de Bits da Cache = Bits dos Dados + Bits das Tags

    2. Encontrando os bits dos dados:
        - 32KB x 8 = 2^15 x 2^3 = 2^18 bits, portanto 2^18 = 262.144 bits para os dados.

    3. Encontrando os bits da tag:
        - A MP tem 16MB = 2^24 células, logo o endereço tem 24 bits.

            - SEPARANDO: do total de 24 bits, no formato [TAG | CONJUNTO | CÉLULA]
                - cada linha tem 8 células de largura -> 3 bits para a célula
                - nº de linhas = 32KB / 8B = 2^12; conjunto de 4 = 2^2 linhas por conjunto, então 2^12 / 2^2 = 2^10 conjuntos -> 10 bits para o conjunto
                - sobram 24 - 10 - 3 = 11 bits para a tag

            - Calculando o total de bits:
                - Dados: 2^18 bits
                - Tags: 11 x (nº de linhas) = 11 x 2^12, pois cada uma das 2^12 linhas guarda uma tag de 11 bits
                - ***Total = 11 x 2^12 + 2^18 = 45.056 + 262.144 = 307.200 bits = 300 Kbits***

- Encontrando o endereço divido na forma (bloco|célula):
    1. MEMORIA PRINCIPAL:
        - Se a MP tem 64MB, ela precisa de 64M endereços para representar 1 célula. Logo 64M pode ser escrito como 2^26, obtendo que precisamos de 26 bits para representar cada célula.

    2. CONJUNTO: 
        - Calculado como total de linhas/nº de linhas no conjunto, para isso precisamos do numero de linhas:
            - LINHAS: Dado por capacidade da cache dividivo pela largura de cada linha, portanto 2^18 (32KB) / 2^7 (16B)  -> 2^11, logo 11 bites para linhas

        - CONJUNTOS: Como a linha é dada por 2^11 e temos 2^2 linhas por bloco, o número de blocos é 2^9, necessitanto de 9 bits para ser representado.
        

    2. CONCLUSÃO:
        - Se no total 26 bits de endereço e 9 são para o conjunto e 5 para o byte logo o formato do endereço é: ***[ 13bits(tag) | 9bits(cojunto) | 4bits(célula) ]***

- Encontrando o endereço dividido na forma (tag | conjunto | célula):

    1. MEMÓRIA PRINCIPAL:
        - O endereço 3FC92B6 tem 7 algarismos hexadecimais. Como cada hexa = 4 bits, o endereço tem 7 x 4 = 28 bits, logo a MP é 2^28 = 256MB.

    2. CÉLULA (byte):
        - Cada linha/bloco tem 32 bytes = 2^5, logo são 5 bits para a célula.

    3. CONJUNTO:
        - Calculado como nº de linhas / nº de linhas por conjunto; para isso preciso do nº de linhas:
            - LINHAS: capacidade da cache / largura da linha = 2^16 (64KB) / 2^5 (32B) -> 2^11, logo 2^11 linhas.
        - CONJUNTOS: como o conjunto é de 4 = 2^2 linhas por conjunto, então 2^11 / 2^2 = 2^9 conjuntos, necessitando de 9 bits.

    4. TAG:
        - Sobra: 28 - 9 (conjunto) - 5 (célula) = 14 bits.

    5. FORMATO: ***[ 14 bits (tag) | 9 bits (conjunto) | 5 bits (célula) ]***

    6. EXTRAINDO O CONJUNTO do endereço 3FC92B6:
        - 3FC92B6 = 0011 1111 1100 1001 0010 1011 0110 (28 bits)
        -  ***Valor binário do conjunto = 010010101***
