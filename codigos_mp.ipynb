## FAZENDO MEMORIA 32X16 COM 4 BANCOS 



```python
from math import log2

TOTAL_POSICOES = 32      
LARGURA_BITS   = 16      
N_BANCOS       = 4       

POSICOES_POR_BANCO = TOTAL_POSICOES // N_BANCOS          

BITS_ENDERECO = int(log2(TOTAL_POSICOES))               
BITS_BANCO    = int(log2(N_BANCOS))                     
BITS_OFFSET   = int(log2(POSICOES_POR_BANCO))           

memoria = [
    [["0"] * LARGURA_BITS for _ in range(POSICOES_POR_BANCO)]
    for _ in range(N_BANCOS)
]


def ler_endereco():
    texto = input(f"Digite o endereço de {BITS_ENDERECO} bits : ")
    endereco = int(texto, 2)
    return endereco


def ler_dado():
    texto = input(f"Digite o dado de {LARGURA_BITS} bits: ")   
    return list(texto)


def localizar(endereco):
   
    banco = endereco//POSICOES_POR_BANCO
    posicao = endereco%POSICOES_POR_BANCO
    return banco, posicao


print("MEMORIA 32X16 com 4 BANCOS")
memoryActive = True

while memoryActive:
    print("Digite W para escrever, R para ler e Q para parar: ")
    acao = input(" ").upper()

    if acao == "W":
        endereco = ler_endereco()
       
        dado = ler_dado()
        
        banco, posicao = localizar(endereco)
        memoria[banco][posicao] = dado
        print(f"salvo no banco {banco}, posição {posicao}.")

    elif acao == "R":
        endereco = ler_endereco()
        
        banco, posicao = localizar(endereco)
        print(f"banco {banco}, posição {posicao} : {''.join(memoria[banco][posicao])}")
        

    elif acao == "Q":
        memoryActive = False

    else:
        print("opção inválida!")
```

## FAZENDO MEMORIA 32M x 16 



```python
from math import log2

TOTAL_ENDERECOS = 2**22
TOTAL_LINHAS = 2**11
TOTAL_COLUNAS = 2**11
LARGURA_BITS = 1
        
BITS_ENDERECOS = int(log2(TOTAL_ENDERECOS))                   


memoria = [
    [["0"] * LARGURA_BITS for _ in range(TOTAL_COLUNAS)] for _ in range (TOTAL_LINHAS)
]


def ler_endereco():
    endereco = int(input(f"Digite o endereço de ( 0 a {TOTAL_ENDERECOS - 1} ) em decimal: "))
    return endereco


def ler_dado():
    texto = input(f"Digite o dado de {LARGURA_BITS} bits: ")   
    return list(texto)


def localizar(endereco):       
    linha, coluna  = divmod(endereco, TOTAL_LINHAS)
    return linha, coluna


print("MEMORIA 4096K x 1")
memoryActive = True

while memoryActive:
    print("Digite W para escrever, R para ler e Q para parar: ")
    acao = input(" ").upper()

    if acao == "W":
        endereco = ler_endereco()
       
        dado = ler_dado()
        
        linha, coluna = localizar(endereco)
        memoria[linha][coluna] = dado
        print(f"salvo na coluna {coluna}, linha {linha}.")

    elif acao == "R":
        endereco = ler_endereco()
        
        linha, coluna = localizar(endereco)
        print(f"  coluna {coluna}, linha {linha} : {''.join(memoria[linha][coluna])}")

    elif acao == "Q":
        memoryActive = False

    else:
        print("opção inválida!")
```

## FAZENDO A MEMÓRIA 128M x 4 COM 2 BANCOS

Ao testar percebi que demorou cerca de 1 min e 30 segundos para rodar, então caso queira testar se sinta a vontade para esperar, pois o código vai rodar e ser completamente funcional.*

```python
from math import log2

TOTAL_ENDERECOS = 2**27
TOTAL_LINHAS = 2**13
TOTAL_COLUNAS = 2**13 
LARGURA_BITS = 4     
N_BANCOS = 2
        

BITS_ENDERECOS = int(log2(TOTAL_ENDERECOS))               
BITS_BANCO  = int(log2(N_BANCOS))                    


memoria = [
    [[["0"] * LARGURA_BITS for _ in range(TOTAL_COLUNAS)] for _ in range (TOTAL_LINHAS)]
    for _ in range(N_BANCOS)
]


def ler_endereco():
    endereco = int(input(f"Digite o endereço de ( 0 a {TOTAL_ENDERECOS - 1} ) em decimal: "))
    return endereco


def ler_dado():
    texto = input(f"Digite o dado de {LARGURA_BITS} bits: ")   
    return list(texto)


def localizar(endereco):
    banco, resto   = divmod(endereco, 2**26)       
    linha, coluna  = divmod(resto, 8192)
    return banco, linha, coluna


print("MEMORIA  COM 2 BANCOS")
memoryActive = True

while memoryActive:
    print("Digite W para escrever, R para ler e Q para parar: ")
    acao = input(" ").upper()

    if acao == "W":
        endereco = ler_endereco()
       
        dado = ler_dado()
        
        banco, linha, coluna = localizar(endereco)
        memoria[banco][linha][coluna] = dado
        print(f"salvo no banco {banco}, coluna {coluna}, linha {linha}.")

    elif acao == "R":
        endereco = ler_endereco()
        
        banco, linha, coluna = localizar(endereco)
        print(f"banco {banco}, coluna {coluna}, linha {linha} : {''.join(memoria[banco][linha][coluna])}")

    elif acao == "Q":
        memoryActive = False

    else:
        print("opção inválida!")
```
